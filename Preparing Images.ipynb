```python
# importing the module
import cv2
  
coordinates_x=[]
coordinates_y = []
# function to display the coordinates of
# of the points clicked on the image
def click_event(event, x, y, flags, params):
 
    # checking for left mouse clicks
    if event == cv2.EVENT_LBUTTONDOWN:
 
        # displaying the coordinates
        # on the Shell
        print(x, ' ', y)
 
        # displaying the coordinates
        # on the image window
        coordinates_x.append(x)
        coordinates_y.append(y)
        font = cv2.FONT_HERSHEY_SIMPLEX
        cv2.putText(img, str(x) + ',' +
                    str(y), (x,y), font,
                    1, (255, 0, 0), 2)
        cv2.imshow('image', img)
 
    # checking for right mouse clicks    
    if event==cv2.EVENT_RBUTTONDOWN:
 
        # displaying the coordinates
        # on the Shell
        print(x, ' ', y)
 
        # displaying the coordinates
        # on the image window
        font = cv2.FONT_HERSHEY_SIMPLEX
        b = img[y, x, 0]
        g = img[y, x, 1]
        r = img[y, x, 2]
        cv2.putText(img, str(b) + ',' +
                    str(g) + ',' + str(r),
                    (x,y), font, 1,
                    (255, 255, 0), 2)
        cv2.imshow('image', img)
 
# driver function
if __name__=="__main__":
 
    # reading the image
    img = cv2.imread('Brain (2).jpg', 1)
 
    # displaying the image
    cv2.imshow('image', img)
 
    # setting mouse handler for the image
    # and calling the click_event() function
    cv2.setMouseCallback('image', click_event)
 
    # wait for a key to be pressed to exit
    k = cv2.waitKey(0) & 0xFF

    if k == 27:  # close on ESC key
        cv2.destroyAllWindows()
 
```

```python
# path

   
# Reading an image in default mode
image = cv2.imread('Brain (1).jpg', 1)
   
# Window name in which image is displayed
window_name = 'Image'
  
# Center coordinates
center_coordinates = (coordinates_x[1],coordinates_y[1] )
 
# Radius of circle
radius = 14
  
# Blue color in BGR   
color = (255, 0, 0)
  
# Line thickness of 2 px
thickness = -1
  
# Using cv2.circle() method
# Draw a circle with blue line borders of thickness of 2 px



TEXT_FACE = cv2.FONT_HERSHEY_DUPLEX
TEXT_SCALE = 0.4
TEXT_THICKNESS = 1
TEXT = "Fp1"
#image = cv2.flip(image, 1)
image = cv2.circle(image, center_coordinates, radius, color, thickness)
text_size, _ = cv2.getTextSize(TEXT, TEXT_FACE, TEXT_SCALE, TEXT_THICKNESS)
text_origin = (coordinates_x[1] - int(text_size[0] / 2), coordinates_y[1] + int(text_size[1] / 2))

cv2.putText(image, TEXT, text_origin, TEXT_FACE, TEXT_SCALE, (255,255,255), TEXT_THICKNESS, cv2.LINE_AA)

cv2.imshow(window_name, image)
k = cv2.waitKey(0) & 0xFF
# Displaying the image
if k == 27:
    # close on ESC key
    cv2.destroyAllWindows()

    


```
